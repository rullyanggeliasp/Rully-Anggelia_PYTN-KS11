## confidence interval

- confidence interval 95% maka z-score 1.90
- standart eror for population proportin : akar dr population proportion +((1-pp)/number of observation))

```python
n = 86+226
pro_fm = 226/n
print(n,pro_fm)
```

```python
stderror_fm = (pro_fm*(1-pro_fm)/n)**(1/2)
stderror_fm
```

```python
z_score = 1.96
low_ci_fm = pro_fm - (z_score*stderror_fm)
up_ci_fm = pro_fm + (z_score*stderror_fm)
print(low_ci_fm,up_ci_fm)
```

```python
print('Margin Errors=',(z_score*stderror_fm))
```

# Hypothesis Testing in Population Means

```python
df = pd.read_csv('nhanes_2015_2016.csv')
df
```

uji apakan rata rat bmi dr laki laki lebih tinggi dr pada bmi perempuan

- h0 = rata rata BMI laki laki tdk lebih tinggi signifikan dr perempuan

- h1 = rata2 BMI laki2 lebih tinggi signifikan dr perempuan

- h0 : meanL <= meanP

- h1: meanL > meanP

- RIAGENDR female = 2 Male=1

```python
females = df[df.RIAGENDR==2]
males = df[df.RIAGENDR==1]
```

```python
#  jumlah data, mean, stdeviasi
n_females = len(females)
mean_females = females['BMXBMI'].mean()
std_females = females['BMXBMI'].std()
print(n_females,mean_females,std_females)
```

```python
# jumlah data, mean, stdeviasi
n_males = len(males)
mean_males = males['BMXBMI'].mean()
std_males = males['BMXBMI'].std()
print(n_males,mean_males,std_males)
```

```python
import statsmodels.api as sm
```

```python
alpha = 0.05 # karena convidence lever 95% maka alpha=0.05
```

```python
bmi_hypo = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='larger')
bmi_hypo
```

```python
if bmi_hypo[1]>alpha:
    print('Accept Null Hypotesis (ho) Mean of Males BMI is not Larger than Female')
else:
    print('Reject Null Hypotesis (h0) Mean of Makes BMI is Larger that Females')
```

## data sma hipotesis berbeda

- uji apakah mean bmi laki2 tidak sama dg mean bmi perempuan
- h0 : meanL = meanP
- h1 : meanL != meanP

```python
bmi_hypo_2s = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='two-sided')
bmi_hypo_2s
```

```python
if bmi_hypo_2s[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```

## HIPOTESIS TESTING POPULATION PROPORTION

```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
```

```python
pro_hypo = sm.stats.proportions_ztest(n*pro_alt,n,pro_null,alternative='larger')
pro_hypo
```

```python
if pro_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```
