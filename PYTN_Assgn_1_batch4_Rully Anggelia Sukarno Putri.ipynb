```python
# merubah kolom boroght menjadi country
df.rename(columns={'borough':'country'}, inplace=True)
df.head()
```

```python
# melakukan slicing terhadap kolom yang dipakai
df_country = df.loc[:,['country','major_category','minor_category','year','value']]
df_country.head()
```

```python
# melihat data kasus crime per country
df_summed = df_country.groupby(['country'], as_index=False)['value'].sum()
df_summed
```

```python
# melihat kasus crime per tahun (majr_category + minor_category)
df_sum_year = df.groupby(['year'], as_index=False)['value'].sum()
df_sum_year
```

```python
# menampilkan line plot
df_sum_year.set_index('year').plot(kind='line', figsize=(8,6), color='red', linewidth='2')
plt.title('London Crime Tahun 2008-2016')
plt.ylabel('Number of Crimes')
plt.xlabel('Years')
plt.show()
```

- #### dari grafik line plot di atas pada tahun 2014 mengalami penurunan sedangkan pada tahun 2016 mengalami kenaikan.

```python
# menampilkan jumlah crime terkecil
df_sum_year.min()
```

```python
# menampilkan jumlah crime terbesar
df_sum_year.max()
```

### Menampilkan data dari country

```python
# melihat area brent
Westminster = df[(df['country']=='Westminster')]
Westminster
```

```python
df_Westminster = Westminster[(Westminster['year'] > 2007)].groupby(['year'])['value'].sum()
df_Westminster
```

```python
# menampilkan Line Plot Crime Country Barnet dari tahun 2008-2016
df_Westminster.plot(kind='line', figsize=(30,5), color='red', linewidth='2')
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('LONDON CRIME 2008 - 2016')
plt.show()
```

- #### jadi dari grafik Line PLot di atas pada tahun 2012 adalah puncak dari jumlah kasus crime di london

```python
# menampilan top 5 country dengan jumlah crime tertinggi
df_summed_sorted = df_summed.sort_values(by='value', ascending=False)
df_summed_sorted.head(5)
```

```python
df_top5 = df_summed_sorted.head(5)
df_top5.set_index('country', inplace = True)
df_top5
```

### Menampikan Area Plot dari top 5 country dengan value crime tertinggi

```python
mpl.style.use(['ggplot'])
df_top5.plot(kind = 'area',
             stacked = False,
             figsize=(12,4),)
```

- #### Jadi dari Area Plot ditunjukkan bahwa kota westminster merupakan kota dengn jumlah kasus kriminal tertinggi

```python
# menampilkan area plot majorr_category pada country westminster
df_Westminster_areaPlot = Westminster[(Westminster['year'] > 2007)].groupby(['major_category','year'])['value'].sum()
df_Westminster_areaPlot.plot(kind='area', figsize=(30,5))
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('Major Category Area Westminster 2008-2016')
plt.show()
```

```python
# menampilkan area plot minor_category pada country westminster
df_Westminster_areaPlot = Westminster[(Westminster['year'] > 2007)].groupby(['minor_category','year'])['value'].sum()
df_Westminster_areaPlot.plot(kind='area', figsize=(30,5))
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('Minor Category Area Westminster 2008-2016')
plt.show()
```

```python
# menampilkan major_category = robbery dari tahun 2008-206
major = df[(df['year'] > 2007)].groupby(['major_category','year'])['value'].sum()
robbery = major['Robbery']
robbery
df_robbery = robbery.to_frame()
df_robbery = df_robbery.rename(columns={'value': 'Robbery'})
df_robbery
```

```python
# menampilkan Major_Category = Burglary dari tahun 2008-2016
burglary = major['Burglary']
burglary
df_burglary = burglary.to_frame()
df_burglary = df_burglary.rename(columns={'value': 'Burglary'})
df_burglary
```

```python
# menggabungkan 2 data frame
df_burglary_robbery = pd.concat([df_robbery, df_burglary], axis=1)
df_burglary_robbery
```

```python
df_burglary_robbery.plot(kind='hist')
plt.title("Histogram of Burglary and Robbery Tahun 2008-2016")
plt.ylabel('number of years')
plt.xlabel('number of crimes')
plt.show()
```

```python
df_bar_sum = df.groupby(['country'], as_index=False)['value'].sum()
df_bar_sum
```

```python
df_bar = df_bar_sum.set_index('country')
df_bar.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country') # add to x-label to the plot
plt.ylabel('Numbor of Crime') # add y-label to the plot
plt.title('Histogram of london crime') # add title to the plot

plt.show()
```

## Pie Chart

```python
df_pie = df.groupby(['country'], as_index=False)['value'].sum()
df_pie = df_pie.sort_values(by='value', ascending=False)
df_pie = df_pie.head(5)
df_pie_pre = df_pie.set_index('country')
```

```python
df_pie
```

```python
# menampilkan pie chart
df_pie_pre['value'].plot(kind='pie',
                            figsize=(5,6),
                            autopct="%1.1f%%",
                            startangle=90,
                            shadow=True);
plt.title('Diagram Pie Chart')
plt.axis('equal')
plt.show()
```
